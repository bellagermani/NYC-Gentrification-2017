```python
import pandas as pd
```

```python
OCCindividuals = pd.read_csv('/Users/isabellagermani/Desktop/MS THSS/data/rawindividualsOCC.csv', index_col = 0)

OCCindividuals = OCCindividuals[['SBA','Age','Hispanic','Race','Year moved In','Reason for Move','Tenure','Immigrant','Monthly Gross Rent','Total Household Income','Rent % Income','Welfare','General Health','Number of persons in household','Gender']].copy()
OCCindividuals
```

```python
SBA = pd.read_csv('/Users/isabellagermani/Desktop/MS THSS/data/SBA.csv', index_col = 0)
SBA.head(5)
```

```python
# checking to make sure right type
print(SBA['SBA'].dtype)
print(OCCindividuals['SBA'].dtype)
```

```python
# checking length
print(len(SBA))
print(len(OCCindividuals))
```

```python
# merging OCCindividuals & SBA on SBA

# LEFT merge 
subanalysis_merged = pd.merge(
    left = OCCindividuals,          
    right = SBA,                
    on = 'SBA',                  
    how = 'left'              
)
subanalysis_merged
```

```python
# double checking length post-merge
len(subanalysis_merged)
```

```python
# constructing displaced binary column
displacement_codes = [16, 18]

subanalysis_merged['Displaced'] = subanalysis_merged['Reason for Move'].isin(displacement_codes).astype(int)

# verification
print('Number Displaced: ', subanalysis_merged['Displaced'].sum())
print('Not displaced: ', len(subanalysis_merged) - subanalysis_merged['Displaced'].sum())
print('Missing Reason for Move: ', subanalysis_merged['Reason for Move'].isna().sum())
```

```python
# constructing gentrified & active gentrified binary columns
gentrification_codes = [3, 6, 10, 12, 13, 14, 17, 19]

# conditions
reason_condition = subanalysis_merged['Reason for Move'].isin(gentrification_codes)
income_condition = subanalysis_merged['Total Household Income'] >= (1.5 * subanalysis_merged['med_hous_inc'])

subanalysis_merged['Gentrified'] = (income_condition).astype(int)
subanalysis_merged['Active Gentrified'] = (reason_condition & income_condition).astype(int)

# verification
print('Number Gentrified (income only): ', subanalysis_merged['Gentrified'].sum())
print('Not Gentrified: ', len(subanalysis_merged) - subanalysis_merged['Gentrified'].sum())
print('Number Active Gentrified (both conditions): ', subanalysis_merged['Active Gentrified'].sum())
print('Not Active Gentrified: ', len(subanalysis_merged) - subanalysis_merged['Active Gentrified'].sum())
```

```python
# constructing hispanic
Hispanic_codes = [2,3,4,5,6,7]

subanalysis_merged['is_hispanic'] = subanalysis_merged['Hispanic'].isin(Hispanic_codes).astype(int)

# verification
print('Number Hispanic: ', subanalysis_merged['is_hispanic'].sum())
print('Not Hispanic: ', len(subanalysis_merged) - subanalysis_merged['is_hispanic'].sum())
```

```python
# constructing races
black_code = [2]
white_code = [1]
asian_codes = [4,5,6,7,8]

subanalysis_merged['is_black'] = subanalysis_merged['Race'].isin(black_code).astype(int)
subanalysis_merged['is_white'] = subanalysis_merged['Race'].isin(white_code).astype(int)
subanalysis_merged['is_asian'] = subanalysis_merged['Race'].isin(asian_codes).astype(int)

# verification
print('Number Black: ', subanalysis_merged['is_black'].sum())
print('Not Black: ', len(subanalysis_merged) - subanalysis_merged['is_black'].sum())
print('Number White: ', subanalysis_merged['is_white'].sum())
print('Not White: ', len(subanalysis_merged) - subanalysis_merged['is_white'].sum())
print('Number Asian: ', subanalysis_merged['is_asian'].sum())
print('Not Asian: ', len(subanalysis_merged) - subanalysis_merged['is_asian'].sum())
```

```python
# constructing immigrant
immigrant_code = [1]

subanalysis_merged['is_immigrant'] = subanalysis_merged['Immigrant'].isin(immigrant_code).astype(int)

# verification
print('Number Immigrants: ', subanalysis_merged['is_immigrant'].sum())
```

```python
# handling strange values
subanalysis_merged = subanalysis_merged[(subanalysis_merged['Total Household Income'] >= 1) & (subanalysis_merged['Total Household Income'] <= 9999997)] # topcode amt
subanalysis_merged = subanalysis_merged[(subanalysis_merged['Monthly Gross Rent'] >= 1) & (subanalysis_merged['Monthly Gross Rent'] <= 5995)] # topcode amt
subanalysis_merged = subanalysis_merged[(subanalysis_merged['Rent % Income'] >= 1) & (subanalysis_merged['Rent % Income'] <= 1010)]
subanalysis_merged = subanalysis_merged[(subanalysis_merged['General Health'] >= 1) & (subanalysis_merged['General Health'] <= 5)]
subanalysis_merged['Tenure'] = subanalysis_merged['Tenure'].replace(9, 0)
subanalysis_merged['Gender'] = subanalysis_merged['Gender'].replace(1, 0) # M
subanalysis_merged['Gender'] = subanalysis_merged['Gender'].replace(2, 1) # F
subanalysis_merged['Welfare'] = subanalysis_merged['Welfare'].replace(2, 0)

# converting rent % income to percentages
temporary = subanalysis_merged['Rent % Income'].astype(str).str.zfill(4)
subanalysis_merged['Rent % Income'] = temporary.str[:3].astype(int) + (temporary.str[3].astype(int) / 10) # to get percent
```

```python
subanalysis_merged = subanalysis_merged[['SBA','Displaced','Gentrified','Active Gentrified','Age','is_hispanic','is_black','is_white','is_asian','is_immigrant','Year moved In','Tenure','Monthly Gross Rent','Total Household Income','Rent % Income','Welfare','General Health','Number of persons in household','Gender']]
```
