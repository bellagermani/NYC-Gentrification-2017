```python
crosswalk = pd.read_csv('https://www2.census.gov/geo/docs/maps-data/data/rel/2010_Census_Tract_to_2010_PUMA.txt', dtype=str)
crosswalk
```

```python
# filtering for NYstate 
Crosswalk = crosswalk[crosswalk['STATEFP'] == '36'].copy()
Crosswalk
```

```python
# filtering for City counties
county = ('005', '047', '061', '081', '085')

CRosswalk = Crosswalk[Crosswalk['COUNTYFP'].isin(county)].copy()
CRosswalk
```

```python
# Full Fips tract code...
CRosswalk['tract_fips10'] = CRosswalk['STATEFP'] + CRosswalk['COUNTYFP'] + CRosswalk['TRACTCE'].copy()
CRosswalk
```

```python
# getting rid of useless columns
CROsswalk = CRosswalk[['PUMA5CE','tract_fips10']].copy()
CROsswalk
```

```python
# dropping leading 0s from PUMA
CROsswalk['PUMA'] = CROsswalk['PUMA5CE'].astype(str).apply(lambda x: x[1:] if x.startswith('0') else x).copy()
CROsswalk['PUMA'].unique()
```

```python
# dictionary for adding in SBA...
sba_to_puma = {
    301 : 3810, # sub-borough area to PUMA (2010)
    302 : 3809, # mn
    303 : 3807,  
    304 : 3808, 
    305 : 3806,
    306 : 3805,
    307 : 3802,
    308 : 3803,
    309 : 3804,
    310 : 3801,
    101 : 3710, # bx
    102 : 3705,
    103 : 3708,
    104 : 3707,
    105 : 3706,
    106 : 3701,
    107 : 3709,
    108 : 3703,
    109 : 3704,
    110 : 3702,
    201 : 4001, # bk
    202 : 4004,
    203 : 4003,
    204 : 4002,
    205 : 4008,
    206 : 4005,
    207 : 4012,
    208 : 4006,
    209 : 4011,
    210 : 4013,
    211 : 4017,
    212 : 4014,
    213 : 4018,
    214 : 4015,
    215 : 4016,
    216 : 4007,
    217 : 4010,
    218 : 4009,
    401 : 4101, # qns
    402 : 4109,
    403 : 4102,
    404 : 4107,
    405 : 4110,
    406 : 4108,
    407 : 4103,
    408 : 4106,
    409 : 4111,
    410 : 4113,
    411 : 4104,
    412 : 4112,
    413 : 4105,
    414 : 4114,
    501 : 3903, # si
    502 : 3902,
    503 : 3901  
}
```

```python
# reversing it bc keys & values are in the wrong places
puma_2_sba = { v:k for (k,v) in sba_to_puma.items()}
puma_2_sba
```

```python
# Adding SBA to crosswalk
# first changing puma column to be same type
CROsswalk['PUMA'] = CROsswalk['PUMA'].astype(int)

CROsswalk['SBA'] = CROsswalk['PUMA'].map(puma_2_sba).copy()
CROsswalk
```

```python
# Lastly, dropping PUMA5CE

CROSSWALK = CROsswalk[['tract_fips10','PUMA','SBA']].copy()
CROSSWALK
```
