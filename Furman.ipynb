```python
# making new small df to hold the 2013-2017 / 2017-2021 vars
cols_to_move = ['SBA','SBA_name','year','aff_30','aff_80','aff_120','vac_rt']

little_df = FURMAN2017[cols_to_move]
little_df

little_df_clean = little_df[little_df['year'].astype(str).str.len() != 4]
little_df_clean
```

```python
little_df_clean = little_df_clean.copy()
```

```python
# taking out % signs & replacing all nans w/ median

# aff_30
little_df_clean['aff_30'] = pd.to_numeric(little_df_clean['aff_30'].str.rstrip('%'), errors='coerce')
median = median_value = little_df_clean['aff_30'].median()
little_df_clean['aff_30'] = little_df_clean['aff_30'].fillna(median).apply(lambda x: f"{x:.1f}")
little_df_clean['aff_30']

# aff_80
little_df_clean['aff_80'] = pd.to_numeric(little_df_clean['aff_80'].str.rstrip('%'), errors='coerce')
median = median_value = little_df_clean['aff_80'].median()
little_df_clean['aff_80'] = little_df_clean['aff_80'].fillna(median).apply(lambda x: f"{x:.1f}")
little_df_clean['aff_80']

# aff_120
little_df_clean['aff_120'] = pd.to_numeric(little_df_clean['aff_120'].str.rstrip('%'), errors='coerce')
median = median_value = little_df_clean['aff_120'].median()
little_df_clean['aff_120'] = little_df_clean['aff_120'].fillna(median).apply(lambda x: f"{x:.1f}")
little_df_clean['aff_120']

# vac_rt
little_df_clean['vac_rt'] = pd.to_numeric(little_df_clean['vac_rt'].str.rstrip('%'), errors='coerce')
median = median_value = little_df_clean['vac_rt'].median()
little_df_clean['vac_rt'] = little_df_clean['vac_rt'].fillna(median).apply(lambda x: f"{x:.1f}")
little_df_clean['vac_rt']
```

```python
# Reformatting as pre/post vars

# aff_30
little_df_clean = little_df_clean.assign(
    aff_30_pre=little_df_clean['aff_30'].where(little_df_clean['year'] == '2013-2017'),
    aff_30_post=little_df_clean['aff_30'].where(little_df_clean['year'] == '2017-2021')
)

# aff_80
little_df_clean = little_df_clean.assign(
    aff_80_pre=little_df_clean['aff_80'].where(little_df_clean['year'] == '2013-2017'),
    aff_80_post=little_df_clean['aff_80'].where(little_df_clean['year'] == '2017-2021')
)

# aff_120
little_df_clean = little_df_clean.assign(
    aff_120_pre=little_df_clean['aff_120'].where(little_df_clean['year'] == '2013-2017'),
    aff_120_post=little_df_clean['aff_120'].where(little_df_clean['year'] == '2017-2021')
)

# vac_rt
little_df_clean = little_df_clean.assign(
    vac_rt_pre=little_df_clean['vac_rt'].where(little_df_clean['year'] == '2013-2017'),
    vac_rt_post=little_df_clean['vac_rt'].where(little_df_clean['year'] == '2017-2021')
)

little_df_clean
```

```python
SURROUNDING_YEARS = little_df_clean[['SBA','aff_30_pre','aff_30_post','aff_80_pre','aff_80_post','aff_120_pre','aff_120_post','vac_rt_pre','vac_rt_post']]
SURROUNDING_YEARS

# getting rid of nans by summing each sba (each var has one numerical and one nan)
SURROUNDING_YEARS = SURROUNDING_YEARS.groupby('SBA').sum().reset_index()
SURROUNDING_YEARS
```
