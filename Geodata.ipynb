```python
import pandas as pd
import geopandas as gpd
```

```python
shape = gpd.read_file('/Users/isabellagermani/Desktop/MS THSS/data/puma_shape/nypuma2010.shp')

cross = pd.read_csv('/Users/isabellagermani/Desktop/MS THSS/data/CROSSWALK.csv', index_col = 0)
# aggregating crosswalk to one row per PUMA
cross_puma = cross.drop_duplicates(subset='PUMA').sort_values('SBA').copy()
cross_puma = cross_puma[['PUMA','SBA']].copy()
print(cross_puma)

sba = pd.read_csv('/Users/isabellagermani/Desktop/MS THSS/data/SBA.csv', index_col = 0)
```

```python
print("Shapefile columns:")
print(shape.columns.tolist())
print(shape['PUMA'].dtype)

print("Crosswalk columns:")
print(cross_puma.columns.tolist())
print(cross_puma['PUMA'].dtype)

print("sba columns:")
print(sba.columns.tolist())
print(sba['SBA'].dtype)
```

```python
# merging w crosswalk
shape['PUMA'] = shape['PUMA'].astype(int)
cross_puma['PUMA'] = cross_puma['PUMA'].astype(int)
sba['SBA'] = sba['SBA'].astype(int)

puma_with_sba = shape.merge(cross_puma, on='PUMA', how='left').sort_values('SBA')
puma_with_sba
```

```python
# merging w sba
sba_geo = puma_with_sba.merge(sba, on='SBA', how='left')
sba_geo

# & dropping puma
cols_to_keep = [col for col in sba_geo.columns if col != 'PUMA']
SBA_GEO = sba_geo[cols_to_keep]
SBA_GEO
```

```python
# Exporting
SBA_GEO.to_file('/Users/isabellagermani/Desktop/MS THSS/data/SBA_GEO.geojson', driver='GeoJSON')
```
