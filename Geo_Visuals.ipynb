```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
geo = gpd.read_file('/Users/isabellagermani/Desktop/MS THSS/data/SBA_GEO.geojson')
geo.head(4)
```

```python
# displacement map
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
geo.plot(column='displaced', 
         ax=ax, 
         legend=True,
         cmap='Reds',
         edgecolor='black',
         linewidth=0.5)
ax.set_title('2017 Spatial Distribution of Displaced Populations', fontsize=16)
ax.set_axis_off()
plt.show()
```

```python
# gentrification map
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
geo.plot(column='PCT_GENTR', 
         ax=ax, 
         legend=True,
         cmap='Blues',
         edgecolor='black',
         linewidth=0.5)
ax.set_title('2017 Spatial Distribution of Gentrifying Populations', fontsize=16)
ax.set_axis_off()
plt.show()
```

```python
# active gentrification map
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
geo.plot(column='PCT_ACTIVE_GENTR', 
         ax=ax, 
         legend=True,
         cmap='Blues',
         edgecolor='black',
         linewidth=0.5)
ax.set_title('2017 Spatial Distribution of Active Gentrifying Populations', fontsize=16)
ax.set_axis_off()
plt.show()
```

```python
# side-by-side of gentrifiers & active gentrifiers

fig, (ax1,ax2) = plt.subplots(1,2, figsize = (20,10))

# gentrifiers
geo.plot(column='PCT_GENTR', 
         ax=ax1, 
         legend=False,
         cmap='Blues',
         edgecolor='black',
         linewidth=0.5)
ax1.set_title('2017 Spatial Distribution of Gentrifying Populations', fontsize=16)
ax1.set_axis_off()

# active gentrifiers 
geo.plot(column='PCT_ACTIVE_GENTR', 
         ax=ax2, 
         legend=False,
         cmap='Blues',
         edgecolor='black',
         linewidth=0.5)
ax2.set_title('2017 Spatial Distribution of Active Gentrifying Populations', fontsize=16)
ax2.set_axis_off()

plt.tight_layout()
plt.show()
```

```python
# community resilience map
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
geo.plot(column='resilience_index', 
         ax=ax, 
         legend=True,
         cmap='Purples',
         edgecolor='black',
         linewidth=0.5)
ax.set_title('2017 Spatial Distribution of Community Resilience', fontsize=16)
ax.set_axis_off()
plt.show()
```

```python
# side-by-side of displacees & gentrifiers

fig, (ax1,ax2) = plt.subplots(1,2, figsize = (20,10))

# incoming Displacees
geo.plot(column='displaced', 
         ax=ax1, 
         legend=False,
         cmap='Reds',
         edgecolor='black',
         linewidth=0.5)
ax1.set_title('2017 Spatial Distribution of Current Displacees', fontsize=16)
ax1.set_axis_off()

# Incoming Active Gentrifiers 
geo.plot(column='PCT_ACTIVE_GENTR', 
         ax=ax2, 
         legend=False,
         cmap='Blues',
         edgecolor='black',
         linewidth=0.5)
ax2.set_title('2017 Spatial Distribution of Active Gentrifiers', fontsize=16)
ax2.set_axis_off()

plt.tight_layout()
plt.show()
```

```python
# side-by-side of rgi & community resilience

fig, (ax1,ax2) = plt.subplots(1,2, figsize = (20,10))

# rgi
geo.plot(column='rent_gap_index', 
         ax=ax1, 
         legend=False,
         cmap='Blues',
         edgecolor='black',
         linewidth=0.5)
ax1.set_title('2017 Spatial Distribution of the Rent Gap', fontsize=16)
ax1.set_axis_off()

# Community resilience 
geo.plot(column='resilience_index', 
         ax=ax2, 
         legend=False,
         cmap='Purples',
         edgecolor='black',
         linewidth=0.5)
ax2.set_title('2017 Spatial Distribution of Community Resilience', fontsize=16)
ax2.set_axis_off()

plt.tight_layout()
plt.show()
```
