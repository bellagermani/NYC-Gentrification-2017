```python
# merging occupied & vacant together to get full NYCHVS dataset
NYCHVS2017 = pd.concat([occupied,vacant], axis = 0, join = 'outer', ignore_index=True)
NYCHVS2017

# sorting by SBA
NYCHVS2017.sort_values(by = ['SBA'], inplace = True)
NYCHVS2017
```

```python
# SEPERATE THING - PCT_GENTR REVISION (Identifying a household as gentrifiers if they make over 1.5x the median household income)
# And ACTIVE_GENTR (a smaller group of gentrifiers defined by income (making 1.5x median income) AND motivation (moved for specfic reasons - see data dictionary))

# loading in SBA to get the med hous inc
SBA = pd.read_csv('/Users/isabellagermani/Desktop/MS THSS/data/SBA.csv')
SBA['SBA'].dtype
SBA['med_hous_inc'].dtype

# mapping med_hous_incomes to SBA so i can match them to the household level SUBSET
sba_income_dict = SBA.set_index('SBA')['med_hous_inc'].to_dict()

# Creating subset to use for calculation of pct_gentr2.0 
SUBSET = NYCHVS2017[['SBA','Year moved In','Total Household Income','Reason for Move']].copy()
SUBSET 

# adding in med_household_income that i calculated in SBA
SUBSET['med_hous_inc'] = SUBSET['SBA'].astype(float).map(sba_income_dict).copy()
SUBSET
```

```python
# now calculating gentr 2.0 & active gentr

# gentrifier
SUBSET['income_ratio'] = SUBSET['Total Household Income'] / SUBSET['med_hous_inc']
SUBSET['gentrifier'] = (SUBSET['income_ratio'] > 1.5).astype(float) # making more than 1.5x more than the median household income

# reason gentrifier (listed one of the reasons for move)
SUBSET['reason_gentrifier'] = (SUBSET['Reason for Move'].isin([3.0, 6.0, 10.0, 12.0, 13.0, 17.0, 19.0]).astype(float))

# ACTIVE gentrifier 
SUBSET['active_gentrifier'] = ((SUBSET['gentrifier'] + SUBSET['reason_gentrifier'] == 2)).astype(float)

# now adding back to the original df to aggregate in the next step
NYCHVS2017['Gentrifier'] = SUBSET['gentrifier']
NYCHVS2017['Active_Gentrifier'] = SUBSET['active_gentrifier']
NYCHVS2017

# & checking if that worked right
print('Total households: ', len(NYCHVS2017))
print('Total Gentrifiers: ', NYCHVS2017['Gentrifier'].sum())
print(f"Percentage: {(NYCHVS2017['Active_Gentrifier'].sum() / len(NYCHVS2017) * 100):.2f}%") 
print('Total Active Gentrifiers: ', NYCHVS2017['Active_Gentrifier'].sum())
print(f"Active Percentage: {(NYCHVS2017['Gentrifier'].sum() / len(NYCHVS2017) * 100):.2f}%")  
```

```python
# adding gentrifiers who've moved in in the last 1 & 5 years

# FOR GENTRIFIERS
# recent gentrifier (5 yrs)
SUBSET['recent_gentrifier'] = ((SUBSET['gentrifier'] == 1) & (SUBSET['Year moved In'] >= 2013) & (SUBSET['Year moved In'] <= 2017))
SUBSET['recent_gentrifier'] = SUBSET['recent_gentrifier'].astype(float)
# new gentrifier (1 yr)
SUBSET['new_gentrifier'] = ((SUBSET['gentrifier'] == 1) & (SUBSET['Year moved In'] >= 2016) & (SUBSET['Year moved In'] <= 2017))
SUBSET['new_gentrifier'] = SUBSET['new_gentrifier'].astype(float)

# FOR ACTIVE GENTRIFIERS
# recent active gentrifier (5 yrs)
SUBSET['recent_active_gentrifier'] = ((SUBSET['active_gentrifier'] == 1) & (SUBSET['Year moved In'] >= 2013) & (SUBSET['Year moved In'] <= 2017))
SUBSET['recent_active_gentrifier'] = SUBSET['recent_active_gentrifier'].astype(float)
# new active gentrifier (1 yr)
SUBSET['new_active_gentrifier'] = ((SUBSET['active_gentrifier'] == 1) & (SUBSET['Year moved In'] >= 2016) & (SUBSET['Year moved In'] <= 2017))
SUBSET['new_active_gentrifier'] = SUBSET['new_active_gentrifier'].astype(float)

# adding them to NYCHVS2017
NYCHVS2017['Recent_Gentrifier'] = SUBSET['recent_gentrifier']
NYCHVS2017['New_Gentrifier'] = SUBSET['new_gentrifier']
NYCHVS2017['Recent_Active_Gentrifier'] = SUBSET['recent_active_gentrifier']
NYCHVS2017['New_Active_Gentrifier'] = SUBSET['new_active_gentrifier']
NYCHVS2017

# checks
print('Total gentrifiers: ', NYCHVS2017['Gentrifier'].sum())
print('Recent gentrifiers: ', NYCHVS2017['Recent_Gentrifier'].sum())
print('New gentrifiers: ', NYCHVS2017['New_Gentrifier'].sum())
print('Active gentrifiers: ', NYCHVS2017['Active_Gentrifier'].sum())
print('Recent active gentrifiers: ', NYCHVS2017['Recent_Active_Gentrifier'].sum())
print('New active gentrifiers: ', NYCHVS2017['New_Active_Gentrifier'].sum())
# noted that reason for move codes are only 2014 on, so there are no active gentrifiers before then
```

```python
# RESTRUCTURING so each obs = 1 sba in 2017 (currently each obs is 1 household)
# making neighborhood-level df & storing new aggregated variables in it

NYCHVS_NEIGHBORHOOD_2017 = NYCHVS2017.groupby('SBA').agg(

    # HOUSING SUPPLY 
    vac_rt = ('Occupancy', lambda x: (x == 3).mean()), # Vacancy Rate: proportion of True values (vacant units) in each neighborhood
    condo = ('Condo/Coop', lambda x: (x == 2).mean()), # Percent condos
    coop = ('Condo/Coop', lambda x: (x == 3).mean()), # Percent coops

    # HOUSING QUALITY
    sound = ('Building Condition', lambda x: (x == 2).mean()), # Percent of buildings in "sound" condition (not dilapidated or deteriorating)
    deter = ('Building Condition', lambda x: (x == 3).mean()), # Percent of deteriorating buildings
    rats = ('Rats', lambda x: (x == 1.0).mean()), # Percent of units with rats present
    r_aches = ('R*aches', lambda x: x.isin([2.0,3.0,4.0]).mean()), # Percent of units with r*aches present

    # ECONOMIC VARIABLES
    
    afford = ('Affordability', lambda x: x.isin([1.0,2.0]).mean()), # Percent of residents who find their unit affordable
    unafford = ('Affordability', lambda x: x.isin([4.0,5.0]).mean()), # Percent of residents who find their unit unaffordable
    rent_burd = ('Rent % Income', lambda x: x[x.between(1.0,1000.0)].median() / 1000), # Median rent-to-income ratio (only accounting for values between 1 & 100%)

    # RESIDENTS
    avg_age = ('Age','mean'), # Avg Age
    howner = ('Tenure', lambda x: (x == 1.0).mean()), # Neighborhood homeownership rate
    immigrant = ('Immigrant', lambda x: (x == 1.0).mean()), # Proportion of immigrant residents
    med_year_moved_in = ('Year moved In', 'median'), # using median because it is probably highly affected by outliers
    med_length_res = ('Year moved In', lambda x: (2017 - x).median().astype(int)), # median length of residence
    recent_moves = ('Year moved In', lambda x: (x >= 2007.0).mean()), # Proportion of residents who moved-in within the last 10 years
    fresh_moves = ('Year moved In', lambda x: (x >= 2012.0).mean()), # Proportion of residents who moved-in within the last 5 years
    displaced = ('Reason for Move', lambda x: x.isin([16.0,18.0]).mean()), # Percent of people who may have been displaced to neighborhood (greater housing affordability/ evicted, displaced, or harassed)
    # NOT USED pct_gentr = ('Reason for Move', lambda x: x.isin([3.0, 6.0, 10.0, 12.0, 13.0, 17.0, 19.0]).mean()), #Percent of people who may have a higher probability of contributing to gentrification of neighborhood (wanted to establish seperate household, better neighborhood services, wanted to own, wanted better quality housing)

    # VACANT UNITS
    rennov = ('Reason Vacant', lambda x: x.isin([2.0,3.0]).mean()), # Percent of vacant units waiting to undergo or are currently undergoing rennovation

    # GENTR 2.O
    total_households = ('Gentrifier', 'count'),
    total_gentrifiers = ('Gentrifier', 'sum'),
    Recent_Gentrifier = ('Recent_Gentrifier', 'sum'),
    New_Gentrifier = ('New_Gentrifier', 'sum'),
    Active_Gentrifier = ('Active_Gentrifier', 'sum'),
    Recent_Active_Gentrifier = ('Recent_Active_Gentrifier', 'sum'),
    New_Active_Gentrifier = ('New_Active_Gentrifier', 'sum')

).reset_index()

# percent of households that are gentrifiers (as aggregated columns)
NYCHVS_NEIGHBORHOOD_2017['PCT_GENTR'] = (NYCHVS_NEIGHBORHOOD_2017['total_gentrifiers'] / NYCHVS_NEIGHBORHOOD_2017['total_households']) * 100
NYCHVS_NEIGHBORHOOD_2017['GENTR'] = (NYCHVS_NEIGHBORHOOD_2017['total_gentrifiers'] / NYCHVS_NEIGHBORHOOD_2017['total_households']) # proportion
# recent
NYCHVS_NEIGHBORHOOD_2017['recent_gentr'] = (NYCHVS_NEIGHBORHOOD_2017['Recent_Gentrifier'] / NYCHVS_NEIGHBORHOOD_2017['total_households']) # proportion
# new
NYCHVS_NEIGHBORHOOD_2017['new_gentr'] = (NYCHVS_NEIGHBORHOOD_2017['New_Gentrifier'] / NYCHVS_NEIGHBORHOOD_2017['total_households']) # proportion
# for actives...
NYCHVS_NEIGHBORHOOD_2017['PCT_ACTIVE_GENTR'] = (NYCHVS_NEIGHBORHOOD_2017['Active_Gentrifier'] / NYCHVS_NEIGHBORHOOD_2017['total_households']) * 100
NYCHVS_NEIGHBORHOOD_2017['ACTIVE_GENTR'] = (NYCHVS_NEIGHBORHOOD_2017['Active_Gentrifier'] / NYCHVS_NEIGHBORHOOD_2017['total_households']) # proportion
# recent
NYCHVS_NEIGHBORHOOD_2017['recent_active_gentr'] = (NYCHVS_NEIGHBORHOOD_2017['Recent_Active_Gentrifier'] / NYCHVS_NEIGHBORHOOD_2017['total_households']) 
# new
NYCHVS_NEIGHBORHOOD_2017['new_active_gentr'] = (NYCHVS_NEIGHBORHOOD_2017['New_Active_Gentrifier'] / NYCHVS_NEIGHBORHOOD_2017['total_households']) 
```

```python
# now deleting columns only used for calculation
NYCHVS_NEIGHBORHOOD_2017.drop(columns = ['total_households','total_gentrifiers','Recent_Gentrifier','New_Gentrifier','Active_Gentrifier','Recent_Active_Gentrifier','New_Active_Gentrifier'], inplace = True)
```
