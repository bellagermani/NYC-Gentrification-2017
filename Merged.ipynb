### Creating indexes

```python
# Community Resilience Index

sb = SBA.copy()

# creating necessary variables - inverted displacement (displacement resistance measure)
sb['displ_resist'] = 1 - sb['displaced']

vars = ['displ_resist','howner','med_length_res','immigrant','Dayc_per_1kppl','MDprog_per_1kppl','relig_per_1kppl']

# percentile ranks 
# normalizing (percentiles)
for var in vars:
    if var in sb.columns:
        sb[var + '_normalized'] = sb[var].rank(pct=True) * 100

# weights
weights = {
    'howner_normalized':0.20,
    'med_length_res_normalized':0.20,
    'immigrant_normalized':0.15,
    'displ_resist_normalized':0.15,
    'Dayc_per_1kppl_normalized':0.10,
    'MDprog_per_1kppl_normalized':0.10,
    'relig_per_1kppl_normalized':0.10
}

# calculating index
SBA['resilience_index'] = (
    sb['howner_normalized'] * weights['howner_normalized'] +
    sb['med_length_res_normalized'] * weights['med_length_res_normalized'] +
    sb['immigrant_normalized'] * weights['immigrant_normalized'] +
    sb['displ_resist_normalized'] * weights['displ_resist_normalized'] +
    sb['Dayc_per_1kppl_normalized'] * weights['Dayc_per_1kppl_normalized'] +
    sb['MDprog_per_1kppl_normalized'] * weights['MDprog_per_1kppl_normalized'] +
    sb['relig_per_1kppl_normalized'] * weights['relig_per_1kppl_normalized']
)

# scaling 0-100
SBA['resilience_index'] = SBA['resilience_index'].rank(pct=True) * 100
SBA['resilience_index'].describe()
```

```python
# Family Index

family_data = SBA[['Dayc_per_1kppl', 'I/Fserv_per_1kppl', 'C/Yserv_per_1kppl']].copy()

# normalizing (percentiles)
family_percentiles = family_data.rank(pct = True) * 100

# averaging 
SBA['family_index'] = family_percentiles.mean(axis = 1)
print(SBA['family_index'].describe())
```

```python
# Amenities
necessary = SBA[['dollars_per_1kppl', 'convenience_per_1kppl', 'laundry_per_1kppl', 'post_per_1kppl', 'Ustor_per_1kppl', 'H/Gstor_per_1kppl']].copy()
unnecessary = SBA[['Members_per_1kppl', 'Arts/E_per_1kppl', 'Hotels_per_1kppl', 'R/Pcent_per_1kppl']].copy()

# normalizing (percentiles)
# necessary
necessary_percentiles = necessary.rank(pct = True) * 100
# unnecessary
unnecessary_percentiles = unnecessary.rank(pct = True) * 100

# averaging 
SBA['necessary_amenities_index'] = necessary_percentiles.mean(axis = 1)
SBA['unnecessary_amenities_index'] = unnecessary_percentiles.mean(axis = 1)

# calculating ratio
SBA['amenity_ratio'] = (SBA['unnecessary_amenities_index'] / SBA['necessary_amenities_index'])
```

```python
print(SBA['necessary_amenities_index'].describe())
print(SBA['unnecessary_amenities_index'].describe())
print(SBA['amenity_ratio'].describe())
```

```python
# Rent gap index (also weighted)

sba = SBA.copy()

# creating necessary variables
sba['not_sound'] = 1 - sba['sound']
sba['condocoop'] = sba['condo'] + sba['coop']
sba['decay'] = sba['deter'] + sba['not_sound']
# adding to SBA
SBA['condocoop'] = sba['condocoop']
SBA['decay'] = sba['decay'] 
SBA['not_sound'] = 1 - sba['sound']

variables = ['avg_surrounding_med_inc', 'Share_Priv_Rec', 'MTAStops_per_sqm', 'Wavg_download_sp', 'condocoop', 'vac_rt', 'decay','med_gro_rnt']

# normalizing (percentiles)
for var in variables:
    if var in sba.columns:
        sba[var + '_normalized'] = sba[var].rank(pct=True) * 100

# POTENTIAL RENT proxy (high proxy = high potential)
potential_rent_proxy = (
    0.30 * sba['avg_surrounding_med_inc_normalized'] + 
    0.25 * sba['Share_Priv_Rec_normalized'] +
    0.20 * sba['MTAStops_per_sqm_normalized'] +
    0.15 * sba['Wavg_download_sp_normalized'] +
    0.10 * sba['condocoop_normalized']
)

# CURRENT GROUND RENT
current_ground_rent = sba['med_gro_rnt_normalized']

# DEVALUATION FACTORS (vars that further supress current ground rent)
devaluation = (
    0.50 * sba['decay_normalized'] +
    0.50 * sba['vac_rt_normalized'] 
)

# CURRENT GROUND RENT ADJUSTED FOR DEVALUATION
cgr_proxy_adjusted = current_ground_rent * (1 - devaluation / 100) # adding devaluation as a discount factor

# final rent gap index calculation
sba['rent_gap_index'] = potential_rent_proxy - cgr_proxy_adjusted

# in percentiles
SBA['rent_gap_index'] = sba['rent_gap_index'].rank(pct=True) * 100
print(SBA['rent_gap_index'].describe())
```

```python
# Displacement pressure index

displacement_data = SBA[['rent_burd', 'pov_rate', 'decay', 'mort_forc']]

# normalizing (percentiles)
displacement_percentiles = displacement_data.rank(pct = True) * 100

# averaging 
SBA['displ_pressure_index'] = displacement_percentiles.mean(axis = 1)
print(SBA['displ_pressure_index'].describe())
```
