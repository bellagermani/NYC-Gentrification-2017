### Dollar Stores

```python
dollar = pd.read_csv('/Users/isabellagermani/Desktop/MS THSS/data/dollarstores_tract10.csv')
type(dollar)
```

```python
# filtering for NYC area & year
counties = ('005','081','047','061','085')

filter1 = dollar['year'] == 2017
filter2 = dollar['tract_fips10'].astype(str).str[0:2] == '36'
filter3 = dollar['tract_fips10'].astype(str).str[2:5].isin(counties)

NYC_Dollar = dollar[filter1 & filter2 & filter3].copy()
NYC_Dollar
```

```python
# Using crosswalk
NYC_Dollar['SBA'] = NYC_Dollar['tract_fips10'].astype(str).map(cross_d).copy()
NYC_Dollar
```

```python
# getting rid of useless cols
NYC_Dollar = NYC_Dollar[['SBA','year','totpop','aland10','count_dollarstores','emps_dollarstores']].copy()
NYC_Dollar.head(5)
```

```python
# renaming
NYC_Dollar = NYC_Dollar.rename(columns = {'totpop':'tot_pop','aland10':'sq_miles','count_dollarstores':'t_dollar_stores','emps_dollarstores':'t_dollar_emps'})
NYC_Dollar.head(5)
```

```python
# aggregating to SBA level (multiple tracts in an SBA)
SBA_Dollar = NYC_Dollar.groupby('SBA').agg({
    'tot_pop':'sum',
    'sq_miles':'sum',
    't_dollar_stores':'sum'
}).reset_index()

SBA_Dollar.head(10)
```

```python
# redoing rate calculations at sba level 
# dollar stores per 1k people
SBA_Dollar['dollars_per_1kppl'] = (SBA_Dollar['t_dollar_stores'] / SBA_Dollar['tot_pop']) * 1000

# dollar stores per square mile
SBA_Dollar['dollars_per_sqm'] = SBA_Dollar['t_dollar_stores'] / SBA_Dollar['sq_miles']
```

```python
# Exporting 
DOLLARSTORES2017 = SBA_Dollar.to_csv('/Users/isabellagermani/Desktop/MS THSS/data/DOLLARSTORES2017.csv')
```

### Broadband Internet Availability, Speed, and Adoption

```python
broadb = pd.read_stata('/Users/isabellagermani/Desktop/MS THSS/data/broadband.dta')
type(broadb)
```

```python
# filtering for NYC area & year
counties = ('005','081','047','061','085')

filter1 = broadb['YEAR'] == 2017
filter2 = broadb['TRACT_FIPS10'].astype(str).str[0:2] == '36'
filter3 = broadb['TRACT_FIPS10'].astype(str).str[2:5].isin(counties)

NYC_broadb = broadb[filter1 & filter2 & filter3].copy()
NYC_broadb
```

```python
# Using crosswalk
NYC_broadb['SBA'] = NYC_broadb['TRACT_FIPS10'].astype(str).map(cross_d).copy()
NYC_broadb.head(5)
```

```python
# getting rid of useless cols
NYC_broadb = NYC_broadb[['SBA','TRACT_FIPS10','AVG_DOWNLOAD_SPEED','AVG_UPLOAD_SPEED']].copy()
NYC_broadb.head(5)
```

```python
# *** creating dict of tract populations from one of the dfs above to add into this one so i can calculate the avg sba broadband speeds using population as a weight ***

# filtering arts again
counties = ('005','081','047','061','085')
filter1 = arts['year'] == 2017.0
filter2 = arts['tract_fips10'].astype(str).str[0:2] == '36'
filter3 = arts['tract_fips10'].astype(str).str[2:5].isin(counties)
ARTS = arts[filter1 & filter2 & filter3].copy()

# creating dict from filtered arts df
sba_dict_ = dict(zip(ARTS['tract_fips10'],ARTS['totpop']))
```

```python
# adding populations into broadband
NYC_broadb['tot_pop'] = NYC_broadb['TRACT_FIPS10'].map(sba_dict_)
NYC_broadb.head(5)
```

```python
# calculating weighted speeds for each tract
NYC_broadb['weighted_avg_download'] = NYC_broadb['AVG_DOWNLOAD_SPEED'] * NYC_broadb['tot_pop']

NYC_broadb['weighted_avg_upload'] = NYC_broadb['AVG_UPLOAD_SPEED'] * NYC_broadb['tot_pop']
NYC_broadb.head(5)
```

```python
# aggregating
broadb2017_aggregated = NYC_broadb.groupby(['SBA']).agg({
    'tot_pop': 'sum',
    'weighted_avg_download': 'sum',
    'weighted_avg_upload': 'sum'
}).reset_index()

broadb2017_aggregated.head(10)
```

```python
# calculating weighted avg speeds for each sba
broadb2017_aggregated['Wavg_download_sp'] = broadb2017_aggregated['weighted_avg_download'] / broadb2017_aggregated['tot_pop']
broadb2017_aggregated['Wavg_upload_sp'] = broadb2017_aggregated['weighted_avg_upload'] / broadb2017_aggregated['tot_pop']
```

```python
# dropping cols
BROADB2017 = broadb2017_aggregated[['SBA','Wavg_download_sp','Wavg_upload_sp']].copy()
BROADB2017.head(5)
```

```python
# exporting
BROADB2017 = BROADB2017.to_csv('/Users/isabellagermani/Desktop/MS THSS/data/BROADB2017.csv')
```
